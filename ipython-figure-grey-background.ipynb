I often use `display(fig)` to embed a matplotlib figure in a notebook, instead of using inline mode.
In this way I can both view the image in an external window (for interactivity: zooming etc), and still also embed it in the notebook.

This worked fine, but since IPython 1.0 I noticed that the embedded figure has a grey background (as you can see below) as is the case in the external window (but not if you save the figure).   
This is a little bit ugly in the notebook, and also was not the case in IPython 0.13.

```python
%matplotlib
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import IPython
from IPython.display import display
```

```python
print "ipython: ", IPython.__version__
print "matplotlib: ", matplotlib.__version__
```

```python
fig, ax = plt.subplots()
ax.plot([0,1,2,3,4,5])
display(fig)
```
